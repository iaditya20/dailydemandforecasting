Creating new dataframe to store all the features on which the Target column depends i.e Correlation with target is more than 0.5

```python
X=data[['Order type A','Banking orders (1)','Urgent Order','Banking orders (2)','Order type C','Order type B',' Non-urgent order']]
```

Creating new data frame to store target column

```python
y=data['Target (Total orders)']
```

Dividing the X & y variable into train and test section

```python

from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
```

No of rows and column of train and test data of X and y

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
fit_model= model.fit(X_train, y_train)
Y_pred=fit_model.predict(X_train)
Y_pred_test=fit_model.predict(X_test)
```

Prediction

```python
Y_pred_test
```

Checking the mean square error of Regression ,Less is the MSE better is the model

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_train,Y_pred))
```

```python
y_test
```

Accuracy Calculation

```python
print('Accuracy of model=',model.score(X_test, y_test)*100," %")
```
